```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

def heart(x, y):
    return (x**2 + y**2 - 1)**3 - x**2*y**3

fig, ax = plt.subplots()

x = np.linspace(-1.5, 1.5, 500)
y = np.linspace(-1.5, 1.5, 500)
X, Y = np.meshgrid(x, y)
Z = heart(X, Y)

ax.contour(X, Y, Z, [0], colors='red')
ax.axis([-1.5, 1.5, -1.5, 1.5])
ax.axis('off')

text = ax.text(0, 0, "", fontsize=8, ha='center')

def update(frame):
    ax.clear()
    ax.contour(X, Y, Z, [0], colors='red')
    ax.axis([-1.5, 1.5, -1.5, 1.5])
    ax.axis('off')
    
    i = int(frame/10)
    message = "Te amo"[:i]
    text.set_text(message)
    
ani = FuncAnimation(fig, update, frames=np.arange(0, 101), interval=50)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

def heart(x, y):
    return (x**2 + y**2 - 1)**3 - x**2*y**3

fig, ax = plt.subplots()

x = np.linspace(-1.5, 1.5, 500)
y = np.linspace(-1.5, 1.5, 500)
X, Y = np.meshgrid(x, y)
Z = heart(X, Y)

ax.contour(X, Y, Z, [0], colors='red')
ax.axis([-1.5, 1.5, -1.5, 1.5])
ax.axis('off')

text = ax.text(0, -0.3, "", fontsize=16, ha='center', color='black')

def update(frame):
    ax.clear()
    ax.contour(X, Y, Z, [0], colors='red')
    ax.axis([-1.5, 1.5, -1.5, 1.5])
    ax.axis('off')
    
    i = int(frame/10)
    message = "Te amo"[:i]
    text.set_text(message)
    
ani = FuncAnimation(fig, update, frames=np.arange(0, 101), interval=50)

plt.show()
```
